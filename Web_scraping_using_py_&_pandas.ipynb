EXTARCT THE DATA BY USING BEAUTIFULSOUP

```python
!pip install beautifulsoup4  --quiet --upgrade
from bs4 import BeautifulSoup
```

```python
doc = BeautifulSoup(page.text, 'html.parser')
```

```python
req_table = doc.find_all("table")[1]
table_titles = req_table.find_all('th')
```

```python
print(table_titles)
```

```python
table_head = [i.text.strip() for i in table_titles]
```

```python
table_head
```

```python

import pandas as pd
```

```python
df = pd.DataFrame(columns=table_head)

df
```

```python
column_data= req_table.find_all('tr')
column_data
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df
```

```python
df.to_csv("project_1.csv", index=False)
```
